```python
# top 10 countries with most CO2 emmissions
# visualising countries data in terms of carbon emmission.
avg_co2_emission_acctoentity = df.groupby('Entity')['Value_co2_emissions_kt_by_country'].mean()
avg_co2_emission_acctoentity.head()

#highest_top5_countries =  Value_co2_emissions_kt_by_country.nlargest(10)
```

```python
top_values= avg_co2_emission_acctoentity.nlargest(5)
```

```python
# highest no of co2 emission 
plt.figure(figsize = (7,5))
sns.barplot(x = top_values.index, y = top_values.values)
plt.xlabel('Entity')
plt.ylabel('Average co2 emission')
plt.title('countries with highest number of co2 emissions')
plt.show()

# lowest no of co2 emission
plt.figure(figsize =(7,5))
lowest_country = avg_co2_emission_acctoentity.nsmallest(5)
sns.barplot(x = lowest_country.index, y = lowest_country.values)
plt.xlabel('Entity')
plt.ylabel('Average co2 emission')
plt.title('countries with the lowest rate of emissions')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming 'df' is already defined and contains the necessary data
conditionIndia = df['Entity'] == 'India'
conditionuk = df['Entity'] == 'United Kingdom'
conditionJapan = df['Entity'] == 'Japan'

# Extracting data for India
yeari = df[conditionIndia]['Year']
Aepi = df[conditionIndia]['Aep']

# Extracting data for the United Kingdom
yearuk = df[conditionuk]['Year']
Aepuk = df[conditionuk]['Aep']

# Extracting data for Japan
yearJ = df[conditionJapan]['Year']
AepJ = df[conditionJapan]['Aep']

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(yeari, Aepi, label='India')
plt.plot(yearuk, Aepuk, label='United Kingdom')
plt.plot(yearJ, AepJ, label='Japan')
plt.legend()
plt.title("Access to electricity: A comparison between India and other countries")
plt.xlabel("Year")
plt.ylabel("Access to electricity (%)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Conditions for each country
conditionIndia = df['Entity'] == 'India'
conditionuk = df['Entity'] == 'United Kingdom'
conditionJapan = df['Entity'] == 'Japan'

# Extracting data for India
yeari = df[conditionIndia]['Year']
Aepi = df[conditionIndia]['Aep']

# Extracting data for the United Kingdom
yearuk = df[conditionuk]['Year']
Aepuk = df[conditionuk]['Aep']

# Extracting data for Japan
yearJ = df[conditionJapan]['Year']
AepJ = df[conditionJapan]['Aep']

# Plotting the data
plt.figure(figsize=(12, 8))  # Increased figure size
plt.plot(yeari, Aepi, label='India', linestyle='-', marker='o', linewidth=2, alpha=0.7)
plt.plot(yearuk, Aepuk, label='United Kingdom', linestyle='--', marker='s', linewidth=2, alpha=0.7)
plt.plot(yearJ, AepJ, label='Japan', linestyle='-.', marker='^', linewidth=2, alpha=0.7)

plt.legend()
plt.title("Access to electricity: A comparison between India and other countries")
plt.xlabel("Year")
plt.ylabel("Access to electricity (%)")
plt.grid(True)  # Adding gridlines
plt.show()
```
